Q1: What is the most important factor in determining survival of the Titanic incident?

The Piegraph below shows the follwoing:
> The data in the set given shows that female passengers have higher rate of survival
> It also shows the if you are seated in higher class then the survival is higher
> I have concatenated the two columns Pclass and Sex to get the concatenated data and plot them togather on a pie chart

As you can see on the piechart below:
> If you are a female in any of the class then you have a higher chance of survival
> Male in any of the class has a lower chance of survival
> Also, survival rate is highest if you are a female sitting in 1st class
> Survival rate is lowest for male sitting in Pclass 2

```python
# Creating a separate datafram by selecting Survived, Pclass and Sex columns
titanic_array_Q1 = titanic_df.iloc[: , [1, 2, 4]]
# Filtering the results by only the ones who survived
titanic_array_Q1 = titanic_array_Q1[titanic_array_Q1['Survived'] == 1]
# Concatenating the two columns to get the combined results
titanic_array_Q1['Pclass-Sex'] = titanic_array_Q1['Pclass'].astype(str) + '-' + titanic_array_Q1['Sex']

# Getting the count to be used on piegraph
class_sex_count = titanic_array_Q1.groupby('Pclass-Sex')['Survived'].count()
# Plotting the piegraph
class_sex_count.plot.pie(autopct = '%1.1f%%', subplots=True,figsize=(10, 10), legend = True)
```

Q2: In the movie, the upper-class passengers were given preference on lifeboats. Does this show in the data?

Yes, this does show in the dataset given.

Bargraph analysis:
> The first bargraph shows that the first class passengers have got the highest survival rate
> There is a clear difference between the classes and the survival however the higher the class does not give gurantee that the survival rate will always be higher than the other. As you can see the Pclass 2nd has got lower survival than that of the 3rd.

Piechart analysis:
> The piechart shows the percentage distribution of survival among the three classes
> The class 1 has got the highest ~40% survival rate. So it can be assumed that the passengers travelling with this class ight have been given preference on lifeboats and hence the higher survival
> The other two classes have got this survival percentage 
Pclass 3 - ~35%
Pclass 2 - ~25%


```python
# Creating a subdataframe to use it for Question 2 by selecting the Survived and Pclass columns
titanic_array_Q2 = titanic_df.iloc[ : , [1, 2]]
# Filtering the results only for the survived passengers
titanic_array_Q2 = titanic_array_Q2[titanic_array_Q2['Survived'] == 1]

# Grouping the Question 2 dataframe for Pclass and then taking its count
# This will be now used in PIechart
class_survived_count = titanic_array_Q2.groupby('Pclass')['Survived'].count()


# Using the subplots function to display two graphs side-by-side
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(15, 5))

# Graph -1 Preparing the countplot to display the count on y-axis
ax = sns.countplot(x = 'Pclass' , data = titanic_array_Q2, ax = axes[0])
ax.set_title('Class vs Survival Analysis')
# Graph - 2 Pichart using the Survival and Class percentage
ax = class_survived_count.plot.pie(autopct = '%1.1f%%', ax = axes[1], legend = True)
# Adding a title to the chart
ax.set_title("Class vs Survival Analysis")

```

Q3: “Women and children first”. Was this the case?


Answer: Yes, this was indeed the case. Please refer to the survival analysis done for female, male and children. Out of total female alomst 76% survived. The same percentage for children and male is 50% and 17% respectively.

This clearly shows that female and children were given a priority and hence they survived in more percentage. This could have been due to the physical strength of male. Male are generally considered as stronger than female and children.


Assumption: This piece of code assumes that anyone under 19 years is a child

```python
# Creating another dataframe to work on specific columns Survival, Sex and Age
titanic_array_Q3 = titanic_df.iloc[ : , [1, 4, 5]]

# This for loop replaces 'female' or 'male' value to 'child' in Sex column if the age is less than 19
for i in range(len(titanic_array_Q3)):
    #print(titanic_array_Q3.loc[i, 'Age'])
    if titanic_array_Q3.loc[i, 'Age'].astype(float) < 19:
        titanic_array_Q3.loc[i, 'Sex'] = 'child'


# Preparing the data for three piecharts

# Preparing the female who survived dataframe
titanic_array_Q3_female = titanic_array_Q3.iloc[: , [0, 1]]
titanic_array_Q3_female = titanic_array_Q3_female[titanic_array_Q3_female['Sex'] == 'female']
# Taking the count for Piechart
female_survived_count_Q3 = titanic_array_Q3_female.groupby('Survived')['Sex'].count()

# Preparing the male who survived dataframe
titanic_array_Q3_male = titanic_array_Q3.iloc[: , [0, 1]]
titanic_array_Q3_male = titanic_array_Q3_male[titanic_array_Q3_male['Sex'] == 'male']
# Taking the count for Piechart
male_survived_count_Q3 = titanic_array_Q3_male.groupby('Survived')['Sex'].count()

# Preparing the male who survived dataframe
titanic_array_Q3_children = titanic_array_Q3.iloc[: , [0, 1]]
titanic_array_Q3_children = titanic_array_Q3_children[titanic_array_Q3_children['Sex'] == 'child']
# Taking the count for Piechart
children_survived_count_Q3 = titanic_array_Q3_children.groupby('Survived')['Sex'].count()

# Using the subplots function to display the piecharts side-by-side
fig, axes = plt.subplots(nrows=1, ncols=3,figsize=(15, 10))


#Pichart using the Survival and Class percentage
# Graph 1 - This displays female percentage who survived and not survived
ax = female_survived_count_Q3.plot.pie(autopct = '%1.1f%%', ax = axes[0], legend = True)
ax.set_title('Female')
ax.legend(['Not Survived', 'Survived'])
# Graph 2 - This displays male percentage who survived and not survived
ax = male_survived_count_Q3.plot.pie(autopct = '%1.1f%%', ax = axes[1], legend = True)
ax.set_title('Male')
ax.legend(['Not Survived', 'Survived'])
# Graph 2 - This displays children percentage who survived and not survived
ax = children_survived_count_Q3.plot.pie(autopct = '%1.1f%%', ax = axes[2], legend = True)
ax.set_title('Children')
ax.legend(['Not Survived', 'Survived'])

```

Q4: Add one other observation that you have noted in the dataset.

Additional Observation: Parent - Child vs Survival Analysis

The pie charts below show that passengers having atleast one parent-child had a higher chance of survival. The passengers with atleast one parent-child has the survival rate of 51% where as the ones without any parent-child has the same rate as 34.

There is a clear difference here. This could be down to the fact that children were given priority and hence atleast one of their parent could have been given priority as well.

This is a very interesting finding and something that confirms the other question partially that children might have been given priority


```python
# Creating a dataframe for passengers with atleast one Parent-Child
# by selecting specific columns (Survived, Pclass and Parch) for our analysis
titanic_array_parch_true = titanic_df.iloc[: , [1, 2, 7]]
# Filtering condition on Parch column to filter the passengers with atleast one parent-child
titanic_array_parch_true = titanic_array_parch_true[titanic_array_parch_true['Parch'].astype(int) > 0]
# Takign the count for the piechart
parchTrue_survived_count = titanic_array_parch_true.groupby('Survived')['Parch'].count()

# Creating a dataframe for passengers without any Parent-Child
# by selecting specific columns (Survived, Pclass and Parch) for our analysis
titanic_array_parch_false = titanic_df.iloc[: , [1, 2, 7]]
# Filtering condition on Parch column to filter the passengers without any parent-child
titanic_array_parch_false = titanic_array_parch_false[titanic_array_parch_false['Parch'].astype(int) == 0]
# Takign the count for the piechart
parchFalse_survived_count = titanic_array_parch_false.groupby('Survived')['Parch'].count()

# Pichart using the Survival and Class percentage
# Using the subplots function to display two plots in one row
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(15, 10))
# Graph 1 - This shows Passengers with atleast one Parent - Child
ax = parchTrue_survived_count.plot.pie(autopct = '%1.1f%%', ax = axes[0], legend = True)
ax.set_title('Passengers with atleast one Parent - Child')
ax.legend(['0 - Not Survived', '1 - Survived'])
# Graph 2 - This shows Passengers without any Parent - Child
ax = parchFalse_survived_count.plot.pie(autopct = '%1.1f%%', ax = axes[1], legend = True)
ax.set_title('Passengers without any Parent - Child')
ax.legend(['0 - Not Survived', '1 - Survived'])
```
